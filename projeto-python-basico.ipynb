```python
# Lógica de Programção

# Passo 0 - Entender o desafio que você quer resolver

# Passo 1 - Percorrer todos os arquivos da pasta base de dados (Vendas)

import os # Biblioteca para trabalhar com locais de arquivos no pc
import pandas as pd # Biblioteca para trabalhar com bases de dados

lista_arquivo = os.listdir('/content/drive/MyDrive/Programacao/curso-basico-python-hashtag/Vendas')
display(lista_arquivo) # Exibindo a lista formatada
```

```python
tabela_compilada = pd.DataFrame() # Criando uma tabela vazia (Dataframe = Tabela)
dataframes = []

# Passo 2 - importar as bases de dados de vendas
for arquivo in lista_arquivo:
  # verificar se tenho Vendas no arquivo
  if 'Vendas' in arquivo:
    # Importar o arquivo
    tabela = pd.read_csv(f'/content/drive/MyDrive/Programacao/curso-basico-python-hashtag/Vendas/{arquivo}')
    dataframes.append(tabela)

# Passo 3 - Tratar / Compilar as bases de dados
tabela_compilada = pd.concat(dataframes, ignore_index=True)
display(tabela_compilada)
```

```python
# Passo 4 - Calcular o produto mais vendido(em quantidade)
tabela_produtos = tabela_compilada.groupby('Produto').sum()
tabela_produtos = tabela_produtos[['Quantidade Vendida']].sort_values(by='Quantidade Vendida', ascending=False)
display(tabela_produtos)
```

```python
# Passo 5 - calcular o produto que mais faturou(em faturamento)
tabela_compilada['Faturamento'] = tabela_compilada['Quantidade Vendida'] * tabela_compilada['Preco Unitario']
tabela_faturamento = tabela_compilada.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[['Faturamento']].sort_values(by='Faturamento', ascending=False)
display(tabela_faturamento)
```

```python
# Passo 6 - Calcular a loja/cidade que mais faturou - criar um gráfico/dashboard
tabela_lojas = tabela_compilada.groupby('Loja').sum()
tabela_lojas = tabela_lojas[['Faturamento']].sort_values(by='Loja', ascending=True)
display(tabela_lojas)
```

```python
import plotly.express as px

grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y='Faturamento')
grafico.show()
```
